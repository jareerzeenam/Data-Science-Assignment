```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Set the styles to Seaborn
sns.set()
# Import the KMeans module so we can perform k-means clustering with sklearn
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('GlobalTemperature.csv')
```

```python
data
```

```python
plt.scatter(data['Year'],data['Apr'])
# Set limits of the axes, again to resemble the world map
plt.xlim(1880,2020)
plt.ylim(-0.50,1)
plt.show
```

```python
x=data.iloc[:,0:3]
```

```python
x
```

```python
kmeans = KMeans(3)
```

```python
kmeans.fit(x)
```

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

```python
data_with_clusters = data.copy()
data_with_clusters['Cluster'] = identified_clusters
data_with_clusters
```

```python
plt.scatter(data_with_clusters['Year'],data_with_clusters['Apr'],c=data_with_clusters['Cluster'],cmap='rainbow')
plt.xlim(1880,2020)
plt.ylim(-0.50,1)
plt.show
```
