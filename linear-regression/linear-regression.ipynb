```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('GlobalTemperature.csv')
```

```python
data
```

```python
plt.scatter(data['Year'],data['Apr'])
# Set limits of the axes, again to resemble the world map
plt.xlim(1880,2020)
plt.ylim(-0.50,1)
plt.show
```

```python
x = sm.add_constant(data['Year'])
results = sm.OLS(data['Apr'],x).fit()
results.summary()
```

```python
plt.scatter(data['Year'],data['Apr'])
yhat = 0.0076*data['Year'] + -14.8392
fig = plt.plot(data['Year'],yhat, lw=4, c='orange',label='regression line')
# Set limits of the axes, again to resemble the world map
plt.xlim(1880,2020)
plt.ylim(-0.50,1)
plt.show
```
